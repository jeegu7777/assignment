```python
def remove_outlier(cars, rating):
    q1 = cars[rating].quantile(0.25)
    q3 = cars[rating].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    #fence_low  = q1-1.5*iqr
    #fence_high = q3+1.5*iqr
    df_out = cars.loc[(cars[rating] > q1) & (cars[rating] < q3)]
    return df_out
```

```python
carnew =remove_outlier(cars, "rating")
```

```python
fig = px.box(carnew['rating'])
fig
```
